## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt = """
     Hello my name is Saurav.
     I love to play valorant.. [laughing]
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

## Advanced examples

```python
text_prompt = """
    कहानियों के माध्यम से बच्चों को अच्छी सीख देने वाली ये
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    WOMAN: I love to go out and meet friends.
    MAN: I love to play valorant.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```
